```python
import numpy as np
from scipy.optimize import curve_fit as cf
import matplotlib.pyplot as plt
import pandas as pd

df2 = pd.read_csv('https://raw.githubusercontent.com/Arhaan/CABKrittikaTutorials/main/Selection_Assignment/data.txt')
xdata = 10**((df2.iloc[:,0])/5 -5)
ydata = df2.iloc[:,1]
def lin_func(x,m,c):
    return m*x + c

plt.scatter(xdata,ydata,s=1)
p_opt, p_cov = cf(lin_func,xdata,ydata)
plt.plot(xdata,lin_func(xdata,*p_opt),label='Best Fit',color='g')
plt.scatter(xdata,ydata,label='OG_Data',s=1,color='darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Fit')
plt.legend()
```
